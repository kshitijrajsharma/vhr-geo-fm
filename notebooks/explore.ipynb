```python
# vhr_datasets = ['spacenet2', 'treesatai', 'flair2', 'dynamic_earthnet', 'spacenet7', 'everwatch', 'nzcattle']
vhr_datasets = ['spacenet2', 'treesatai', 'flair2', 'dynamic_earthnet', 'spacenet7']

```

```python
import glob
import os 
repo_path = "/Users/krschap/code/personal/GEO-Bench-2-Leaderboard/results/*/results_and_parameters.csv"
files = glob.glob(repo_path)

df_list = []
for f in files:
    df = pd.read_csv(f)
    df['submission'] = os.path.basename(os.path.dirname(f)) # hash
    df_list.append(df)

df_all = pd.concat(df_list, ignore_index=True)
```

```python
df_all
```

```python
df_vhr = df_all[df_all['dataset'].isin(vhr_datasets)]
df_mean = df_vhr.groupby(['backbone', 'submission', 'dataset'])['test metric'].mean().unstack().reset_index()

df_mean['mean_score'] = df_mean[vhr_datasets].mean(axis=1) * 100 
for col in vhr_datasets:
    if col in df_mean:
        df_mean[col] = df_mean[col] * 100


df_final = df_mean.sort_values(by='mean_score', ascending=False).reset_index(drop=True)
df_final.insert(0, 'Rank', df_final.index + 1)

cols = ['Rank', 'backbone', 'submission', 'mean_score'] + [d for d in vhr_datasets if d in df_final.columns]
df_final = df_final[cols]
```
